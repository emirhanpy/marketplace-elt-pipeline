## TRANSFORM PHASE

```python
import pandas as pd
```

```python
original_df = pd.read_csv('trendyol_mobile_phones.csv')
original_df
```

```python
def transform_from_csv(df):
    
    # Changed data type of the 'RATING SCORE'.
    # Renamed the 'RATING SCORE' to 'RATING_SCORE'.
    df = df.rename(columns={'RATING SCORE': 'RATING_SCORE'})
    df['RATING_SCORE'] = df['RATING_SCORE'].astype(float)
    df['RATING_SCORE'] = df['RATING_SCORE'].fillna(0)
    
    # Assigned '0' to NaN values.
    df['RATING'] = df['RATING'].fillna(0)
    df['RATING'] = df['RATING'].replace(r'[\(\)]', '', regex=True).astype(int)
    
    # Missing 'RATING' values have been filled with the average 'RATING' values of the corresponding 'BRAND'.
    brand_avg_rating = df[df['RATING'] != 0].groupby('BRAND')['RATING'].mean()
    rating_zero = df['RATING'] == 0
    df.loc[rating_zero, 'RATING'] = df.loc[rating_zero, 'BRAND'].map(brand_avg_rating)
    df['RATING'].fillna(0, inplace=True)
    
    # Replaced the character data type for changing the data type and then missing 'PRICE' values have been filled with '0'.
    df['PRICE'] = df['PRICE'].replace({' TL': '', ',': ''}, regex=True)
    df['PRICE'] = df['PRICE'].fillna(0)
    df['PRICE'] = df['PRICE'].str.replace('.', '', regex=False).astype(int)
    
    # Replaced the character data type for changing the data type and then missing 'COUNT OF FAVORITE' values have been filled with '0'. 
    # Dropped the df rows of 'COUNT OF FAVORITE' higher than 250000.
    # Renamed the 'COUNT OF FAVORITE' to 'FAVORITES'.
    df = df.rename(columns={'COUNT OF FAVORITE': 'FAVORITES'})
    df['FAVORITES'] = df['FAVORITES'].str.replace('B kişi favoriledi!', '', regex=False)
    df['FAVORITES'] = df['FAVORITES'].astype(str)
    df['FAVORITES'] = df['FAVORITES'].apply(lambda x: 0 if 'Son 3 günde' in x else x)
    df['FAVORITES'] = df['FAVORITES'].replace('nan', '0')
    df['FAVORITES'] = df['FAVORITES'].str.replace(' kişi favoriledi!', '', regex=False)
    df['FAVORITES'] = df['FAVORITES'].astype(float).round(decimals=2)
    df['FAVORITES'] = df['FAVORITES'] * 1000
    df['FAVORITES'] = df['FAVORITES'].fillna(0)
    df = df[df['FAVORITES'] < 250000.0]
    
    # Dropped the df rows of 'PRICE' smaller than 500 TRY and higher than 120000 TRY.
    df = df[df['PRICE'] >= 500]
    df = df[df['PRICE'] <= 120000]
    
    # Dropped the df rows of 'PRODUCT' Which 'PRODUCT' is empty.
    df = df.dropna(subset=['PRODUCT'])
    
    # Set the data type of the 'BRAND' column to string.
    df['BRAND'] = df['BRAND'].astype(str)
    
    # Created three new columns: 'PRICE_TRY', 'PRICE_DOLLAR', and 'PRICE_EURO', corresponding to 'PRICE', and then dropped the 'PRICE' column."
    usd_to_try = 33.8
    euro_to_try = 37.5
    df['PRICE_TRY'] = df['PRICE']
    df['PRICE_DOLLAR'] = df['PRICE'] / usd_to_try
    df['PRICE_EURO'] = df['PRICE'] / euro_to_try
    df = df.drop(columns=['PRICE'])
    
    # Removed the least known brands and those that are not brands."
    drop_brands = ['AGM Mobile', 'JAXEL', 'NOTHING', 'OVİON', 'Mipo', 'Hikıng', 'Hiking', 'ULEFONE', 'Trident', 'ouktiel', 'Wiko', 'iflytek', 'MAPLE', 'SEVEN7 MOBİLE', 'Gigaset', 'WORKMATE',
               'Jarvisen', 'Gurlek Group','AsGlobal', 'Seven7', 'İtel', 'FİNEBULE', 'infiniti', 'Axe Tools', 'Zebra', 'PREMIER', 'Chainway', 'mukay', 'Smartbery', 'GENERAL', 'KEEP IN TR',
               'Holographic', 'Urfakent', 'SenalStore', 'Bade', 'Nikadu', 'Nokian', 'Genel Markalar', 'DIGIPHONE', 'FİTGO', 'ENİGMA', 'EMO', 'İXTECH', 'Daylux', 'Smart Tech', 'Bade Shop', 
               'BB Mobile', 'TEMUR AKSESUAR', 'Axel', 'DODGE', 'WildTech', 'NEW', 'Kelepir Fırsatlar', 'Mobile', 'inteknoloji']
    df = df[~df['BRAND'].isin(drop_brands)]
    
    return df
```

```python
df = pd.read_csv('trendyol_mobile_phones.csv')
transformed_df = transform_from_csv(df)
```

```python
transformed_df.to_csv('transformed_data.csv', index=False)
transformed_df
```

## DATA ANALYTICS & DASHBOARD PHASE

```python
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('Trendyol_Mobile_Phones.db')
table_name = 'MOBILE_PHONES'
```

```python
transformed_df.to_sql(table_name, conn, if_exists='replace', index=False)
```

```python
query_statement = f"""
SELECT BRAND,PRODUCT,PRICE_TRY FROM MOBILE_PHONES
GROUP BY PRODUCT
ORDER BY PRICE_TRY DESC
LIMIT 10

"""
query_output = pd.read_sql(query_statement, conn)
print(query_output)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='PRICE_TRY', y='PRODUCT',hue='BRAND', data=query_output, palette='deep')
plt.legend(title='Marka', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

plt.title('TOP 10 Most Expensive Products')
plt.xlabel('PRICE (TRY)')
plt.ylabel('PRODUCT')
plt.show()
```

```python
query_statement_2 = f"""
SELECT BRAND,PRODUCT,PRICE_TRY FROM MOBILE_PHONES
GROUP BY PRODUCT
ORDER BY PRICE_TRY ASC
LIMIT 10

"""
query_output_2 = pd.read_sql(query_statement_2, conn)
print(query_output_2)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='PRICE_TRY', y='PRODUCT',hue='BRAND', data=query_output_2, palette='tab10')
plt.legend(title='Marka', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

plt.title('TOP 10 Most Affordable Products')
plt.xlabel('PRICE (TRY)')
plt.ylabel('PRODUCT')
plt.show()
```

```python
query_statement_3 = f"""
SELECT BRAND, 
ROUND(AVG(PRICE_TRY), 0) AS AVG_PRICE_TRY 
FROM MOBILE_PHONES
GROUP BY BRAND
HAVING COUNT(PRODUCT) > 5
ORDER BY AVG_PRICE_TRY DESC
"""
query_output_3 = pd.read_sql(query_statement_3, conn)
print(query_output_3)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='AVG_PRICE_TRY', y='BRAND', data=query_output_3, palette='tab10')
plt.title('AVG Prices by Brand')
plt.xlabel('AVG PRICE (TRY)')
plt.ylabel('BRAND')
plt.show()
```

```python
query_statement_4 = f"""
SELECT BRAND, 
SUM(FAVORITES) AS TOTAL_FAVORITES
FROM MOBILE_PHONES
GROUP BY BRAND
HAVING COUNT(PRODUCT) > 50
ORDER BY TOTAL_FAVORITES DESC
LIMIT 10
"""
query_output_4 = pd.read_sql(query_statement_4, conn)
print(query_output_4)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='TOTAL_FAVORITES', y='BRAND', data=query_output_4, palette='tab10')
plt.title('Total Favorites by Brand')
plt.xlabel('TOTAL FAVORITES (MILLION)')
plt.ylabel('BRAND')
plt.show()
```

```python
query_statement_5 = f"""
SELECT BRAND, RATING_SCORE
FROM MOBILE_PHONES
GROUP BY BRAND
HAVING COUNT(PRODUCT) > 50
ORDER BY RATING_SCORE DESC
LIMIT 5         
"""
query_output_5 = pd.read_sql(query_statement_5, conn)
print(query_output_5)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='BRAND', y='RATING_SCORE', data=query_output_5, palette='tab10')
plt.title('TOP 5 Rating Scores by Brand')
plt.xlabel('BRAND')
plt.ylabel('RATING_SCORE')
plt.show()
```

```python
query_statement_6 = f"""
SELECT BRAND, 
(ROUND(SUM(RATING), 0) / AVG(RATING_SCORE) ) AS RATING_SCORE_PERFORMANCE 
FROM MOBILE_PHONES
GROUP BY BRAND
HAVING COUNT(PRODUCT) > 50
ORDER BY RATING_SCORE_PERFORMANCE DESC
LIMIT 10
"""
query_output_6 = pd.read_sql(query_statement_6, conn)
print(query_output_6)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='BRAND', y='RATING_SCORE_PERFORMANCE', data=query_output_6, palette='tab10')
plt.title('Rating Score Perfomance by Brand')
plt.xlabel('BRAND')
plt.ylabel('RATING SCORE PERFORMANCE')
plt.show()
```

```python
query_statement_7 = f"""
SELECT BRAND, PRODUCT, 
(ROUND(SUM(RATING), 0) / AVG(RATING_SCORE) ) AS RATING_SCORE_PERFORMANCE 
FROM MOBILE_PHONES
GROUP BY PRODUCT
ORDER BY RATING_SCORE_PERFORMANCE DESC
LIMIT 10
"""
query_output_7 = pd.read_sql(query_statement_7, conn)
print(query_output_7)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='RATING_SCORE_PERFORMANCE', y='PRODUCT',hue='BRAND', data=query_output_7, palette='deep')
plt.title('Rating Score Performance by Products')
plt.xlabel('RATING SCORE PERFORMANCE')
plt.ylabel('PRODUCTS')
plt.show()
```

```python
query_statement_8 = f"""
SELECT BRAND, PRODUCT, (RATING / PRICE_TRY) AS VALUE_FOR_MONEY
FROM MOBILE_PHONES
ORDER BY VALUE_FOR_MONEY DESC
LIMIT 10             
"""
query_output_8 = pd.read_sql(query_statement_8, conn)
print(query_output_8)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='VALUE_FOR_MONEY', y='PRODUCT',hue='BRAND', data=query_output_8, palette='deep')
plt.title('Value for Money Products by Brand')
plt.xlabel('VALUE for MONEY')
plt.ylabel('PRODUCT')
plt.show()
```

```python
query_statement_9 = f"""
SELECT BRAND, PRODUCT, PRICE_TRY
FROM MOBILE_PHONES
WHERE PRICE_TRY = (SELECT MIN(PRICE_TRY)
                   FROM MOBILE_PHONES AS mp
                   WHERE mp.BRAND=MOBILE_PHONES.BRAND
                   HAVING COUNT(PRODUCT) > 30)
OR PRICE_TRY = (SELECT MAX(PRICE_TRY)
                FROM MOBILE_PHONES AS mp2
                WHERE mp2.BRAND=MOBILE_PHONES.BRAND
                HAVING COUNT(PRODUCT) > 30)
ORDER BY BRAND
"""
query_output_9 = pd.read_sql(query_statement_9, conn)
print(query_output_9)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='PRICE_TRY', y='PRODUCT',hue='BRAND', data=query_output_9, palette='tab10')
plt.title('The Highest and Lowest Priced Products by Brand')
plt.xlabel('PRICE (TRY)')
plt.ylabel('PRODUCT')
plt.show()
```

```python
query_statement_10 = f"""
WITH BrandCount AS (
    SELECT BRAND, COUNT(PRODUCT) AS TOTAL_PRODUCT
    FROM MOBILE_PHONES
    GROUP BY BRAND
),
TotalCount AS (
    SELECT COUNT(PRODUCT) AS TOTAL_PRODUCTS
    FROM MOBILE_PHONES
)
SELECT 
    b.BRAND,
    b.TOTAL_PRODUCT,
    (b.TOTAL_PRODUCT * 100.0 / t.TOTAL_PRODUCTS) AS MARKET_SHARE
FROM BrandCount b, TotalCount t
ORDER BY MARKET_SHARE DESC
"""
query_output_10 = pd.read_sql(query_statement_10, conn)
print(query_output_10)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='TOTAL_PRODUCT', y='BRAND', data=query_output_10, palette='deep')
plt.title('Total Products by Brand')
plt.xlabel('TOTAL PRODUCT')
plt.ylabel('BRAND')
plt.show()
```

```python
top_n = 10
top_brands = query_output_10.nlargest(top_n, 'MARKET_SHARE')

brand = top_brands['BRAND']
market_share = top_brands['MARKET_SHARE']

plt.figure(figsize=(12, 8))
plt.pie(market_share, labels=brand, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('deep'))
plt.title(f'TOP {top_n} Brands by Market Share')
plt.axis('equal')
plt.show()
```

```python
query_statement_11 = f"""
SELECT BRAND, PRODUCT, PRICE_TRY, MAX(RATING) AS MAX_RATING
FROM MOBILE_PHONES
GROUP BY PRODUCT
HAVING MAX_RATING
ORDER BY MAX_RATING DESC
LIMIT 10
"""
query_output_11 = pd.read_sql(query_statement_11, conn)
print(query_output_11)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='MAX_RATING', y='PRODUCT',hue='BRAND', data=query_output_11, palette='tab10')
plt.title('TOP 10 Most Rated Products by Brand')
plt.xlabel('RATING')
plt.ylabel('PRODUCT')
plt.show()
```
